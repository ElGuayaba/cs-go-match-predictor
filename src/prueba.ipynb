```python
import pandas as pd

total_data = pd.read_csv('../data/interim/test_data.csv')
team_data = pd.read_csv('../data/raw/team_data.csv')
```

```python
total_data.head(10)
```

```python
def bring_team_id(team_name):
    # Since some teams are uppercase and the database is all lowercase, we need lower the name
    lower_team_name = team_name.lower() 
    result = team_data.loc[team_data['team_name'].str.lower() == lower_team_name, 'team_id'].values[0]
    return result
```

```python
example = total_data.head(20)
example
```

```python
for _,row in example.iterrows():
    example.loc[_,'team_name'] = row.lower()
```

```python
example['team_name'] = example['team_name'].apply(lambda x: x.lower())

example['team_id'] = [team_data.loc[team_data['team_name'] == x, 'team_id'] for x in example['team_name']]
example
```
